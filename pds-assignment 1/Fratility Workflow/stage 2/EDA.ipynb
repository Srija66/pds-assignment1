```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


new_data = pd.read_csv(r"C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\Fratility Workflow\stage 1\Processed_data.csv")
print(new_data.describe())

#plotting Histogram
plt.figure(figsize=(10, 8))
sns.histplot(new_data['Height'], bins=10,color='red',edgecolor='black')
plt.title("Histogram")
plt.savefig("images/hist_height.png")
plt.show()

# plotting Scatterplot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Height', y='Grip strength_new', data=new_data,color='blue',edgecolor='black')
plt.title("Scatter plot")
plt.savefig("images/scatter_height.png")
plt.show()
```
