```python
# Visualization 1

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage1_DP\new_data.csv')

lunch = data['lunch']
average_scores = data['average score']

# Create a bar chart
plt.figure(figsize=(10, 8))
result = data.groupby('lunch')['average score'].mean()
result.plot(kind='bar', color='orange')

# adding labels and title
plt.xlabel('lunch')
plt.ylabel('Avg Score')
plt.title('Bar Chart of lunch vs. Avg Score')

# Saving the figure
plt.savefig(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage2_Visu\Visualizations\bar_chart.png', bboxes_inches='tight')
plt.show()
```
