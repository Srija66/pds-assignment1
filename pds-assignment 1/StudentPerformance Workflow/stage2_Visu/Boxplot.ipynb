```python
#Visualization 2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage1_DP\new_data.csv')

# Assuming dataset has columns 'race/ethnicity' and 'Average Score'
plt.figure(figsize=(10, 8))
sns.boxplot(x='race/ethnicity', y='average score', data=data, palette='pastel')
plt.xlabel('race/ethnicity')
plt.ylabel('Avg Score')
plt.title('Box Plot of race/ethnicity vs. Average Score')
plt.xticks(rotation=90, horizontalalignment='center')

# Save the boxplot
plt.savefig(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage2_Visu\Visualizations\box_plot.png', bbox_inches='tight')
# Show the figure
plt.show()
```
