```python
#Visualization 4

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage1_DP\new_data.csv')

# considering gender and average score
x_data = data['gender']
avg_scores = data['average score']

figure, axis = plt.subplots(figsize=(8, 6))

# histogram for gender and average score
plt.hist([avg_scores[x_data == 'male'], avg_scores[x_data == 'female']], 
         bins=10, alpha=0.5, color=['green', 'yellow'], label=['male', 'female'])

plt.xlabel('Avg Score')
plt.ylabel('Frequency')
plt.title('Histogram of gender vs. Average Score')
plt.legend()

plt.savefig(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage2_Visu\Visualizations\histogram.png', bbox_inches='tight')
plt.show()
```
