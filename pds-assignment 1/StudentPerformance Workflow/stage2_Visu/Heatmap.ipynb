```python
# Visualization 3

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage1_DP\new_data.csv')

# Assuming your dataset has columns 'test preparation course' and 'Average Score'
data['test preparation course'] = data['test preparation course'].map({'completed': 1, 'none': 0})  

corr_matrix = data[['test preparation course', 'average score']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title(' Heatmap of test preparation course vs. Average Score')

plt.savefig(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage2_Visu\Visualizations\correlation_matrix_heatmap.png', bbox_inches='tight')
plt.show()
```
