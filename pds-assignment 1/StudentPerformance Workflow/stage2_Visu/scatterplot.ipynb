```python
# visualization 5

import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage1_DP\new_data.csv')

# considering level of education and average score
degree = data['parental level of education']
avg_scores = data['average score']

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(degree, avg_scores, alpha=0.6,color='red')
plt.xlabel('degree')
plt.ylabel('Avg Score')
plt.title('Scatter Plot of degree vs. Average Score')

# Save the scatterplot
plt.savefig(r'C:\Users\MANOJ\OneDrive\Documents\pds-assignment 1\StudentPerformance Workflow\stage2_Visu\Visualizations\scatter_plot.png', bbox_inches='tight')
plt.show()
```
